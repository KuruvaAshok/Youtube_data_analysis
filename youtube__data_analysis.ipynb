```python
sns.barplot(x='p_year',y='likes',data=video,estimator=len)
```

```python
import ipywidgets as widgets
```

```python
a1=widgets.Dropdown(
    options=video['channel_title'].unique(),
    value='TheEllenShow',
    description='Channel:',
    disabled=False,
)

def graph(k):
  nbc=video[video['channel_title']==k]
  lk=nbc['likes'].sum()
  dl=nbc['dislikes'].sum()
  cc=nbc['comment_count'].sum()
  plt.pie(x=[lk,dl,cc],labels=['likes','disklike','comment_count'],autopct='%1.1f%%')
  plt.show()
```

```python
widgets.interact(graph,k=a1)
```

```python
video['Genre'].value_counts()
```

```python
a1=widgets.Dropdown(
    options=['likes', 'views','comment_count'],
    value='views',
    description='x-axis:',
    disabled=False,
)
def graph(k):
  plt.figure(figsize=(20,4))
  sns.barplot(y=k,x='Genre',data=video,estimator='sum')
  plt.show()
```

```python
widgets.interact(graph,k=a1)
```

```python
#Splitting the data in two parts TRAIN & TEST
from sklearn.model_selection import train_test_split
```

**Model Building**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict([[40000,50,700,24]])
```

```python
lr.score(x_test,y_test)
```
